```python
solution={}
def r(n):   #  对于长度n的价钱
    
    candidates=[]  # 字典 候选的
    
    for i in range(1,n):
        candidates.append((r(i) + r(n-i),i))  # 切分成i段和n-i段  r(i)+r(n-i)是切分后总价
    
    candidates.append((price[n],0))   # 不切分时的价钱
    max_price,split_point = max(candidates,key=lambda x:x[0])
    
    global solution # 全局变量
    solution[n]=(split_point,n-split_point)
        
    return max_price
```

```python
%%time
r(12)
```

```python
def not_cut(n): return n == 0  # left n=0 时 ，并没有切断  # 如果n==0,return空值
```

```python
def parse_solution(target_length, revenue_solution):
    
    left, right = revenue_solution[target_length]
    
    if not_cut(left): return [right]
    
    return parse_solution(left, revenue_solution) + parse_solution(right, revenue_solution)
```

```python
parse_solution(37,solution)
```

```python
from datetime import  datetime
import time
```

```python
# decorator python 装饰器
```

```python
def func_1(n):
    
    begin = datetime.now()
    time.sleep(0.1)
    print('used time = {}'.format(datetime.now()-begin))
    
    return n    
```

```python
def func_2(n):

    begin = datetime.now()
    sum_ = 0
    for i in range(n**n):
        sum_+=1
    print('used time = {}'.format(datetime.now()-begin))
    
    return n
```

```python
func_1(0)
```

 ## python 是一个可以面向函数编程的语言 （函数可以作为变量，可以作为返回值）

```python
def get_func_time(func):
    def _wrap(n):
        begin = datetime.now()
        result = func(n)
        print('used time = {}'.format(datetime.now()-begin))
        
        return result 
    return _wrap
```

```python
get_1_with_time = get_func_time(func_1)
```

```python
get_1_with_time = get_func_time(func_2)
```
